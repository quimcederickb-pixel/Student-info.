```python
#PASS OR FAIL code


while True:

  try:

      score = float(input("(type a letter to end) Enter the student's exam score: "))

      if score >= 75:
          print("Passed")
          continue
      else:
          print("Failed")
          continue
  except ValueError:
      print("Please enter a valid numerical score.")
      decision = input("If you wish to end code, please type END, if not, type anything else")
      if decision == "END":
          break
      elif decision == "end":
          break
      else:
          continue
```

```python
#GRADE code


def calculate_grade(score):
    if 90 <= score <= 100:
        return 'A'
    elif 80 <= score <= 89:
        return 'B'
    elif 70 <= score <= 79:
        return 'C'
    else:
        return 'F'

while True:
  student_name = input("Enter student's name: ")
  score_input = input("Enter student's score: ")

  try:
      student_score = float(score_input)
      if 0 <= student_score <= 100:
          grade = calculate_grade(student_score)
          print(f"\n{student_name}'s grade is: {grade}")
          decision = input("To end the code, please type END, if not, type anything else")
          if decision == "END":
              break
          elif decision == "end":
              break
          else:
              continue
              print()
              print()
      else:
          print("Error: Score must be between 0 and 100.")
  except ValueError:
      print("Error: Invalid score input. Please enter a number.")
      continue

```
