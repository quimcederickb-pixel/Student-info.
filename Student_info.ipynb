```python
import time

#Main loop
while True:
  print("Hello there!")
  print("My name is info. bot!")
  name = input("Please input your name: ")
  print(f"Hello there {name}")

  while True:
    age_str = (input("If I may ask, what is your age? "))
    try:

      age = int(age_str)
      if age  < 1:
        print("That is not a valid age, please try again")
        continue

      break
    except ValueError:
      print("That is not a valid age, please try again")
      continue

  if age >= 18:
    print("You're quite old!")
  elif age <= 17:
    print("You're quite young!")

  course = input("Please input your Pisay course: ")
  print()
  print("I see")
  print()
  time.sleep(2)
  print(f"So, {name} of age {age} is taking the course {course}")
  time.sleep(0.5)
  print(f"thats great {name}!")
  print()
  time.sleep(0.5)
  print(f"See you next time! Goodluck on your {course} course!")
  print()
  time.sleep(3)

  print("Review:")
  print(f"You're name is {name}")
  print(f"You're age is {age}")
  print(f"You're course is {course}")

  time.sleep(1)

  final = input("Is the info correct? (say Y or N)")
  if final == "Y":
    print("Yipeeeeeee")
    break
  elif final == "N":
    print("Oh man, please input you're info again :(")
    print()
    print()
    print()
    print()
    continue



```
